```python
from kafka import KafkaProducer
import os
import time
```

```python
# Configuration Kafka
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',  # Adresse du broker Kafka
    value_serializer=lambda v: v.encode('utf-8')) # Sérialisation des messages
```

```python
# Paramètres
topic = 'test_stream4'
batch_size = 1000  
sleep_interval = 3
hdfs_file_path = r'C:\projet_spark\Project Data\HDFS_DataCenters_logs\DataCenter2\node_logs\hadoop-hdfs-secondarynamenode-mesos-01.log'  # Chemin vers le fichier HDFS
```

```python
def send_batch(batch, producer, topic):
    """Envoie un lot de messages à Kafka."""
    for message in batch:
        try:
            producer.send(topic, value=message)
        except Exception as e:
            print(f"Erreur lors de l'envoi de la ligne : {message}\n{e}")
```

```python
def process_hdfs_file(file_path, batch_size, sleep_interval):
    """Lit un fichier volumineux par lots et envoie les données à Kafka."""
    batch = []
    line_count = 0

    try:
        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip()
                batch.append(line)
                line_count += 1

                # Envoyer le lot lorsque la taille atteint le batch_size
                if len(batch) >= batch_size:
                    send_batch(batch, producer, topic)
                    print(f"{line_count} lignes envoyées.")
                    batch.clear()
                    
                    
                    # Simuler un envoi en temps réel avec une pause
                    time.sleep(sleep_interval)

            # Envoyer le dernier lot s'il y a des lignes restantes
            if batch:
                send_batch(batch, producer, topic)
                print(f"{line_count} lignes envoyées (dernier lot).")

    except Exception as e:
        print(f"Erreur lors du traitement du fichier : {e}")
    finally:
        producer.close()

# Exécution
process_hdfs_file(hdfs_file_path, batch_size,sleep_interval)
```
